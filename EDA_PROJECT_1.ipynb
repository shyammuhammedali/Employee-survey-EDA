```python
#gender distribution

sns.countplot(x='Gender',data=df)
plt.title("gender distribution")
plt.show
```

```python
#martial Status
sns.countplot(x='MaritalStatus',data=df)
plt.title("MaritalStatus")
plt.show
```

```python
sns.histplot(df["Age"],kde=True)
plt.title("age")
plt.show()
```

```python
#TrainingHoursPerYear

sns.histplot(df['TrainingHoursPerYear'], kde=True)
plt.title("TrainingHoursPerYear")
plt.show()
```

```python
#corelation

#select the numerical data

num_df = df.select_dtypes(include=['float64','int64'])

#increase the size of plot

plt.figure(figsize=(15,10))
#heatmap

sns.heatmap(num_df.corr(), cmap='coolwarm', annot=True)
plt.title('correlation of numerical data')
plt.show
```

```python
#realtionship between experience and job satisfaction

sns.scatterplot(x='Experience', y='JobSatisfaction',data=df)
plt.title('Experience and Job Statisfaction')
plt.show()
```
